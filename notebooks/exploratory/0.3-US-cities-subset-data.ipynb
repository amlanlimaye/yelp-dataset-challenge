```python
US_business = business.loc[business['state'].isin(['AZ', 'NV', 'NC', 'PA', 'WI', 'IL', 'SC', 'OH'])]
```

```python
US_tip = tip.loc[tip['business_id'].isin(US_business.business_id.unique())]
```
